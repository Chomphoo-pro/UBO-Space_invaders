```python
# import tkinter as tk
from tkinter import *
import tkinter as tk


#
#contient le squelette de la fenetre + frame
#
class SpaceInvaders(object):
    '''
    Main Game class
    '''
    
    #initialise SpaceInvaders
    def __init__(self):
        self.root = tk.Tk()#créer la fenetre
        self.root.title("Space Invaders")#titre de la fenetre
        self.frame = tk.Frame(self.root) #porte une frame
        self.frame.pack(side="top", fill="both") #affiche la frame
        self.game = Game(self.frame) #créer le jeux
       
        
    #play lance le jeux et affiche la fenetre
    def play(self):
        self.game.start_animation()
        self.root.mainloop()

        
#
#créer la frame
#
class Game(object):
    def __init__(self, frame):
        self.frame = frame
        self.fleet = Fleet()
        self.defender = Defender()
        self.height = 600
        self.width = self.fleet.get_width()
        self.canvas = tk.Canvas(self.frame, width = 1200, height = self.height, bg = 'black')
        self.canvas.pack()
        self.defender.install_in(self.canvas)
        self.fleet.install_in(self.canvas)
        
    def animation(self):
        pass
    
    def start_animation(self):
        self.move_aliens_fleet()
        
    
    def move_bullets(self):
        pass
    
    def move_aliens_fleet(self):
        self.fleet.move_in(self.canvas)
        self.canvas.after(1000, lambda: self.move_aliens_fleet())
    
       
        

        

#
#Defender permet de déplacer la barre
#
class Defender(object):
    def __init__(self):
        self.width = 20
        self.height = 20
        self.move_delta = 20
        self.x = 0
        self.y = 0
        self.id = None
        self.max_fired_bullets = 8
        self.fired_bullets = []
    
    
    
    
    #accesseur
    def get_width(self):
        return self.width
    
    
    def get_height(self):
        return self.height
    
    
    def get_move_delta(self):
        return self.move_delta
    
    
    def get_x(self):
        return self.x
    
    
    def get_y(self):
        return self.y
    
    
    def get_id(self):
        return self.id
    
    
    def get_max_fired_bullets(self):
        return self.max_fired_bullets
    
    
    def get_fired_bullets(self):
        return self.fired_bullets
    
    
    
    #mutateur
    def set_width(self, v):
        self.width = v
        
        
    def set_height(self, v):
        self.height = v

        
    def set_move_delta(self, v):
        self.move_delta = v
        
        
    def set_x(self, v):
        self.x = v
        
        
    def set_y(self, v):
        self.y = v
        
        
    def set_max_fired_bullets(self, v):
        self.max_fired_bullets = v
    
    
    def set_id(self, v):
        self.id = v
        
    
    def set_fired_bullets(self, v):
        self.fired_bullets = v
        
        
        
        
    #créer l'objet Defender
    def install_in(self, canvas):
        canvas_width = canvas['width'] #largeur du canvas
        canvas_width = int(canvas_width) #convertie en int
        canvas_height = canvas['height'] #hauteur du canvas
        canvas_height = int(canvas_height) #convertie en int
        defender_width = self.get_width()#recupere la propriété largeur de defender
        defender_height = self.get_height() #recupere la propriété hauteur de defender
        defender_x = (canvas_width+defender_width)/2 # sur l'axe X le defender est au milieu
        defender_y = (canvas_height-defender_height) # sur l'axe Y le defender est visible tout en bas
        
        self.set_x(defender_x) #modifie la propriété X de defender
        self.set_y(defender_y) #modifie la propriété Y de defender
        
        x1 = defender_x #points du rectangle en haut
        y1 = defender_y #points du rectangle à gauche
        x2 = defender_x + defender_width #points du rectangle en bas
        y2 = defender_y + defender_height #points du rectangle à droite
        
        canvas.focus_set()#le canvas est active afin de récupérer les entrées
        
        
        def events(event): #Action du defender Partie CONTROL
            if (event.keysym == "Left"): #touche fleche de gauche detecte
                dx = -1 #modifie delta x en négatif
                self.move_in(canvas, dx) #déplace le carré à gauche
            elif (event.keysym == "Right"): #touche fleche de droite detecte
                dx = 1 #modifie delta x en positif
                self.move_in(canvas, dx) #déplace le carré à droite

            elif (event.keysym == "space"): #touche espace detecte
                self.fire(canvas) #lance une boulette
            else:
                pass
                #print(event.keysym)
            

        self.set_id(canvas.create_rectangle(x1, y1, x2, y2, fill='white')) #créer le defender
        canvas.bind("<Key>",events) #detecte les touches

        
    

            
    def move_in(self, canvas, dx):
        canvas_width = canvas['width'] #largeur du canvas
        canvas_width = int(canvas_width) #convertie en int
   
        defender_width = self.get_width()#recupere la propriété largeur de defender
        defender_x = self.get_x()#recupere la propriété X de defender
        move_delta = self.get_move_delta()#recupere la propriété move_delta de defender
        
        moving = move_delta * dx #calcul du deplacement de pixel
        
        #future déplacement à gauche du defender
        condition1 = 0 <= defender_x+moving #si le defender ne dépasse pas le bord gauche
        condition2 = defender_x+moving+defender_width <= canvas_width #si le defender ne depasse pas le bord droit
        
        if (condition1 and condition2):
            self.set_x(defender_x+moving)
            canvas.move(self.id, moving, 0) #deplacement du defender en X
        #else:
            #print("x: "+str(defender_x)+"/"+str(self.get_x())+ " - moving: " + str(moving)+ " - dx: "+str(dx)+" - c1: "+str(condition1)+" - c2: "+str(condition2))
            #print("canvas width: "+str(canvas_width))

            
    
    def fire(self, canvas):#Tire une bullette
        lesBullets = self.fired_bullets
        nbrBullets = len(lesBullets)
        maxBullets = self.max_fired_bullets
        if (nbrBullets < maxBullets):
            bullet = Bullet(self)
            lesBullets.append(bullet)
            bullet.install_in(canvas)

                
 
            
        
        
        
    
#  
#Bullet tir des boules en gérant un max
#
class Bullet(object):
    def __init__(self, shooter):
        self.radius = 10
        self.color = "red"
        self.speed = 4
        self.id = None
        self.shooter = shooter
    
    def install_in(self, canvas):
        self.y = self.shooter.y - self.shooter.height
        self.x = self.shooter.x
        self.y = self.y
        
        x = self.x
        y = self.y
        
        self.colision = False
        color = self.color
        r = self.radius
        self.test = 0
        self.id = canvas.create_oval(x, y + 5, x+r, y + 5 +r, fill=color)
        self.move_in(canvas)
        
    
                
                
                
                
    def move_in(self, canvas):
        self.y -= self.speed
        y = self.y
        x = self.shooter.x
        r = self.radius
        
        aliensId = canvas.find_withtag("alien")
        if (self.y > 0 - r*2 and self.colision == False): #deplace en haut
            
        
            
            overlapped = canvas.find_overlapping(x, y, x+r, y + 5 +r)

            #overlapped = overlapped[0] #premier élément
            for id in overlapped:
                if (id in aliensId):
                    self.colision = True
                    #print(id)
                    #a = canvas.find_withtag(id)
                    #print(a)
                    #a.touched_by(canvas, self.id)
                    canvas.delete(id)
                    

                    #if(coord_y >= self.y):
            canvas.move(self.id, 0, -self.speed)
            canvas.after(19, lambda: self.move_in(canvas))
        else: # supprimer quand il arrive en haut
            canvas.delete(self.id)
            self.shooter.fired_bullets.pop()
        
        

        
# 
#Alien instancie 1 alien sous forme d'image
#
class Alien(object):
    def __init__(self):
        self.id = None
        self.alive = True
        self.move_delta = 5
        


    def touched_by(self, canvas, projectile):
        x = projectile.x
        y = projectile.y
        
        if (self.alive):
            self.pim = PhotoImage(file="explosion.gif")
            self.explosion = canvas.create_image(self.x, self.y, image=self.pim, tags="explosion")
            canvas.after(150, lambda: self.touched_by(canvas, projectile))
            self.alive = False
        else:
            self.explosion = canvas.find_withtag("explosion")
            canvas.delete(self.explosion)
            canvas.delete(self.id)
    
    def install_in(self, canvas, x, y, image, tag):
        self.pim = PhotoImage(file=image)
        self.id = canvas.create_image(x, y, image=self.pim, tags=tag)
        

    def move_in(self, canvas, dx, dy):
        movex = self.move_delta * dx
        movey = self.move_delta * dy
        
        #if (self.animation_deplacement == 0):
       #     self.animation_deplacement = 1
       #     self.pim = PhotoImage(file="alien_movement")
       #     self.install_in()
       # else:
       ##     self.pim = PhotoImage(file="alien")
       #     self.animation_deplacement = 0
            
        canvas.move(self.id, movex, movey)
        #canvas.after(1000, lambda: self.move_in(canvas, dx, dy))
        
        
        
#
#Fleet est une flotte d’Alien et son déplacement
#
class Fleet(object):
    def __init__(self):
        self.aliens_lines = 5
        self.aliens_columns = 10
        self.aliens_inner_gap = 20
        self.alien_x_delta = 5
        self.alien_y_delta = 15
        fleet_size = self.aliens_lines * self.aliens_columns
        self.aliens_fleet = [None]* fleet_size
        
    def get_width(self):
        pass
        
        
        
    def install_in(self, canvas):
        self.animation_deplacement = 0
        lines = self.aliens_lines
        columns = self.aliens_columns
        gapx = 0
        gapy = 0
        L = 73 #largeur
        H = 53 #hauteur
        i = 0
        
        for dy in range(lines):
            for dx in range(columns):
    
                alien = Alien()
                alien.install_in(canvas, (dx+1)*L+gapx, (dy+1)*H+gapy, "alien.gif", "alien")
                self.aliens_fleet[i] = alien
                i += 1
                gapx += self.aliens_inner_gap / 2
            gapx = 0
            gapy += self.aliens_inner_gap / 2

    
    def move_in(self, canvas):
        lines = self.aliens_lines
        columns = self.aliens_columns
        dx = self.alien_x_delta
        dy = self.alien_y_delta
        all_rect_ids = canvas.find_withtag("alien")
 
        x1,y1,x2,y2 = canvas.bbox("alien")
        dy = 0
        i = 0
        for l in range(lines):
            for c in range(columns):
                if (x2 > int(canvas["width"])):
                    print("x2")
                    self.alien_x_delta = -5
                    self.aliens_fleet[i].move_in(canvas, self.alien_x_delta, self.alien_y_delta)
                
                elif (x1 < 0):
                    self.alien_x_delta = 5
                    print("x1")
                    self.aliens_fleet[i].move_in(canvas, self.alien_x_delta, self.alien_y_delta)
                else:  

                    self.aliens_fleet[i].move_in(canvas, self.alien_x_delta, 0)
                i += 1
                
                
                
        
        
        
    def manage_touched_aliens_by(self, canvas, defender):
        return 0
    
    
        
    
    
spaceInvaders = SpaceInvaders()
spaceInvaders.play()
```
